```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
time = np.array([0.00, 21.63, 44.73, 92.68, 164.97, 212.72])
Z = np.array([5.6, 6.8, 7.8, 9.6, 11.8, 13.0])
Z_squared = Z ** 2

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(time, Z, 'o-', label="Z vs time")
plt.xlabel("time (hour)")
plt.ylabel("Z (cm)")
plt.title("Z vs time")

plt.subplot(1, 2, 2)
plt.plot(time, Z_squared, 'o-', label="Z^2 vs time", color='orange')
plt.xlabel("time (hours)")
plt.ylabel("Z^2 (cm²)")
plt.title("Z^2 vs time")

plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = linregress(time, Z_squared)
print(f"y = {slope:.2f}x + {intercept:.2f}")
```
