```python
import matplotlib.pyplot as plt
import numpy as np
```

# Problem 2A

```python
rho_L = 992.3 # kg / m^3
C_L = rho_L * 1000 * (1/67.5) # mol / m^3
C_L = C_L / 1000 # mol / L

p_A = [1e-2, 3e-2, 5e-2, 7e-2, 1e-1, 1.1e-1, 1.2e-1, 1.3e-1, 1.4e-1, 1.5e-1, 1.6e-1]
x_A = [2.4e-4, 7.19e-4, 1.15e-3, 1.64e-3, 2.34e-3, 2.58e-3, 2.81e-3, 3.06e-3, 3.28e-3, 3.52e-3, 3.78e-3]
C_A_L = [C_L * frac for frac in x_A]

plt.plot(p_A, C_A_L)
plt.title("Equilibrium Line")
plt.ylabel("Concentration of ClO_2 in Liquid Phase")
plt.xlabel("Pressure (atm)")

print(f"slope: {C_A_L[0]/p_A[0]}")
```

# Problem 2B

```python
m_list = [0] * len(p_A)

for i in range(len(p_A)):
    m_list[i] = C_A_L[i] / p_A[i]

m = np.average(m_list)

print(m_list)
print("The average equilibrium distribution coeffcient is {}".format(m))
```

# Problem 2C (IGNORE FOR GRADING)

```python
k_x = 1.0 # gmol/m^2 * s
k_L_list = [0] * len(p_A)

for i in range(len(p_A)):
    k_L_list[i] = k_x / C_A_L[i]

k_L = np.average(k_L_list)

print(k_L_list)
print("The average k_L value is {}".format(k_L))

# don't need to grade this, just wanted to see
plt.title("Curiosity Plot: k_L vs C_A_L")
plt.plot(C_A_L, k_L_list)
plt.xlabel("Concentration of A in liquid phase")
plt.ylabel("k_L")
plt.grid(True)
```

# Problem 2e (IGNORE FOR GRADING)

```python
k_G = 0.010 # mol/m^2 * s
k_y_list = [k_G * pressure for pressure in p_A]
x_A_bulk = 0.0004

y_A = [0] * len(k_y_list)
for i in range(len(k_y_list)):
    y_A[i] = -(k_y_list[i] / k_L_list[i]) * x_A_bulk + .023

print(y_A)

plt.plot(x_A, y_A, color='b')

```

# Bonus A

```python
x_A = np.linspace(0.01, 0.99, 100)
W_min = x_A * np.log(1 / x_A) + (1 - x_A) * np.log(1 / (1 - x_A))

plt.plot(x_A, W_min, color='b')
plt.title('Minimum Work of Separation')
plt.xlabel('Mole Fraction A')
plt.ylabel("W_min / RT_0 n_F")
```

# Bonus B

```python
y_A_product1 = 0.98
y_B_product2 = 0.98
W_min_b = x_A * np.log(y_A_product1 / x_A) + (1 - x_A) * np.log(y_B_product2 / (1 - x_A))

plt.plot(x_A, W_min, label='Perfect Separation (a)')
plt.plot(x_A, W_min_b, label='98% A to Product 1 (b)', color='r')
plt.title('Minimum Work of Separation')
plt.xlabel('Mole Fraction of A')
plt.ylabel("W_min / RT_0 n_F")
plt.legend()
```

```python
y_A_product1_c = 0.9
y_B_product2_c = 0.9
W_min_c = x_A * np.log(y_A_product1_c / x_A) + (1 - x_A) * np.log(y_B_product2_c / (1 - x_A))

plt.plot(x_A, W_min, label='Perfect Separation (a)', color='b')
plt.plot(x_A, W_min_b, label='98% A to Product 1 (b)', color='r')
plt.plot(x_A, W_min_c, label='90% A in Product 1, 90% B in Product 2 (c)', color='g')
plt.title('Minimum Work of Separation', fontsize=14)
plt.xlabel('Mole Fraction of A')
plt.legend()
```
